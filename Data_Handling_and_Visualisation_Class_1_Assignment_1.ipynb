```python
import matplotlib.pyplot as plt
```

```python
num_categories = df['Category'].nunique()
print(f"Number of unique categories: {num_categories}")
```

```python
category_counts = df['Category'].value_counts()
print(category_counts)
```

```python
category_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Number of Orders by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
region_counts = df['Region'].value_counts()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)

plt.title('Proportion of Orders by Region')
plt.axis('equal')  # Makes the pie chart a circle
plt.show()
```

```python
avg_category_sales = df.groupby('Category')['Sales'].mean().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(8, 5))
avg_category_sales.plot(kind='bar', color='cornflowerblue', edgecolor='black')

plt.title('Average Sales per Order by Category')
plt.xlabel('Category')
plt.ylabel('Average Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract year from the order date
df['Year'] = df['Order Date'].dt.year

# Group by year and sum sales
yearly_sales = df.groupby('Year')['Sales'].sum()

# Plot line graph
plt.figure(figsize=(8, 5))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-', color='tomato')

plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(yearly_sales.index)
plt.tight_layout()
plt.show()
```
